```python
#  ify the black and white colour
# 0-180-->black
# 181-255-->white
thresh,binary_image=cv2.threshold(img_gray,180,255,cv2.THRESH_BINARY)        #180-->black clr,255-->white
cv2.imshow('BINARY IMAGE',binary_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# THRESH_OTSU-->is assign a default threshold value for getting a better image
thresh,binary_image=cv2.threshold(img_gray,10,255,cv2.THRESH_OTSU)        
cv2.imshow('BINARY IMAGE',binary_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt               #we got a bluish shade
plt.imshow(img)
```

```python
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)        #we convert the value into RGB5
plt.imshow(con_img)
```

```python
# in this image we can see that the values of y axis is incremented from upside
```

```python
# rectagle
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.rectangle(conv_img,pt1=(120,10),pt2=(200,70),color=(255,231,123),thickness=2)      #pt1=(120,10)-->left corner point, pt2=(200,70)-->right bottom point
plt.imshow(conv_img)
```

```python
# circle
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.circle(conv_img,center=(160,40),radius=40,color=(0,0,255),thickness=2)
plt.imshow(conv_img)
```

```python
# put text
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.putText(conv_img,text='CUTE TEDDY',org=(50,140),fontFace=cv2.FONT_HERSHEY_TRIPLEX,fontScale=1,color=(0,0,255),thickness=1)
plt.imshow(conv_img)
```

```python
# thickness=thickness of word
# fontscale=size of text
```
