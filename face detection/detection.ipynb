```python
import cv2
img=cv2.imread(r'C:\Users\Luminar Technolab\Desktop\Mubashira DL_projects\face detection\im.jpg')   
img
```

```python
import matplotlib.pyplot as plt               #we got a bluish shade
plt.imshow(img)
```

```python
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
face_cascade=cv2.CascadeClassifier(r'C:\Users\Luminar Technolab\Desktop\Mubashira DL_projects\face detection\haarcascade_frontalface_default.xml')
face_cascade
```

```python
img_gray=cv2.cvtColor(con_img,cv2.COLOR_BGR2RGB)
img_gray
```

```python
face=face_cascade.detectMultiScale(img_gray,minNeighbors=10)  #minNeighbors=6-->when we check the image,that img should detect minimum 6 timem
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(0,0,225),1)  
```
